```python
import requests
import pandas as pd
import re
import datetime
```

```python
#Hier bilde ich aus der Geschäftsnummer-ID das Jahr des Geschäft. In der Regel ist es sinnvoll Funktionen nach den Imports aufzulisten.
def trennen(elem):
    elem = elem.split(".")[0]
    elem = elem.replace('="', "")
    return elem
```

```python
#Das ist im Grunde genommen ein erster Daten import
#feststellen, was das aktuelle Jahr ist, damit in der Liste sämtliche Geschäfte seit 1997 mit dem Jahr versehen werden können
#mit firstyeahr und lastyeahr soll später die Zeitspanne angegeben werden, in der die Geschäfte ausgewertete werden
date = datetime.date.today()
year_akt = date.strftime("%Y")
firstyear = 20
lastyear = 21
```

```python
#Und hier der eigentlicher Daten import.
path_postulate = "postulate.csv"
df_postulate = pd.read_csv(path_postulate, delimiter=';')
```

Liste der Postulate etc. downloaden unter: https://www.ag.ch/grossrat/grweb/de/196/Gesch%C3%A4fte?FrmEntity=grweb.modules.akt.GrProz&FrmRequest=Search&FrmProperty=21d0f0ef601efca72de8b86d83e7d1b4

```python
#Diesen ganzen Abschnitt braucht es dank der obigen Funktioniert nicht. Es ist sinnvoller, mit .apply zu arbeiten.
#mit year_start wird das Startjahr angegeben, um danach durch den Datensatz zu gehen, und bis zum aktuellen Jahr alle Geschäfte mit dem Jahr zu versehen. 
#Weil der Datensatz ab 1997 vorhanden ist, bleibt die Variable besgtehen!

year_start = 1997
while year_start < int(year_akt)+1:
    #der Searchstring sucht z.B. nach [9] [7][.], also genau 97.xxx danach wird 97 in 1997 umgewandelt und bis zum aktullen Jahr hochgezählt
    searchstring = pd.to_datetime(year_start, format='%Y')
    searchstring = searchstring.strftime("%y")
    regex = r"["+str(searchstring[0])+"]["+str(searchstring[1])+"][.]"
    #print(regex)
    for elem in df_postulate.index:
        #print(elem)
        if df_postulate["Ges.-Nr."].str.contains(regex)[elem]:
            df_postulate.iloc[elem]["Jahr"] = year_start
            print(searchstring)
            print(year_start)
        else:
            pass

    year_start+=1
```

```python
#Hier wird die ganze Spalte verareitet und eine neue Spalte kreiert.
df_postulate['Jahre'] = df_postulate['Ges.-Nr.'].apply(trennen)
```
