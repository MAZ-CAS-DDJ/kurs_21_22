# Pivotieren

```python
table = pd.pivot_table(df, values='interessenbindung_art', index=['parlamentarier_anzeige_name'],
                    columns=['organisation_interessengruppe1'], aggfunc="count")
```

```python
table.fillna(0)
```

# Exportieren von Beziehungen

```python
from scipy import sparse
sA = sparse.csr_matrix(table.fillna(0).values)
```

```python
sA
```

```python
G = nx.algorithms.bipartite.matrix.from_biadjacency_matrix(sA)
G.edges(data=True)
```

```python
partei_names = {}
partei_attr = {}
for count,row in enumerate(table.index):
    partei = df[df["parlamentarier_anzeige_name"]==row].iloc[0]["parlamentarier_partei"] #df[df["parlamentarier_anzeige_name"] == 'Addor, Jean-Luc'].iloc[0]["parlamentarier_partei"]
    partei_attr[count] = {"partei": partei}
    partei_names[count] = row
interessen_attribute = {}
interessen_names = {}
for count,row in enumerate(table.columns):
    interessen_attribute[count+len(partei_attr)] = {"interesse":row}
    interessen_names[count+len(partei_names)] = row
partei_attr.update(interessen_attribute)
partei_names.update(interessen_names)
```

```python
nx.set_node_attributes(G, partei_attr)
G = nx.relabel_nodes(G, partei_names)
nx.write_gexf(G,"politiker_und_organisationen.gexf")
```

## Reduzieren auf nur eine Art von Knoten: Politiker

```python
partei_names = {}
partei_attr = {}
for count,row in enumerate(table.index):
    partei = df[df["parlamentarier_anzeige_name"]==row].iloc[0]["parlamentarier_partei"] #df[df["parlamentarier_anzeige_name"] == 'Addor, Jean-Luc'].iloc[0]["parlamentarier_partei"]
    partei_attr[count] = {"partei": partei}
    partei_names[count] = row
politiker = bipartite.weighted_projected_graph(G, list(partei_names.values()))
```

```python
politiker.edges(data=True)
```

```python
nx.write_gexf(politiker,"politiker.gexf")
```
