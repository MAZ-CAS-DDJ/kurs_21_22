Wie ihr seht, sind das ziemlich viele Dokumente. Wie sie alle einlesen und zu jedem Dokumente strukturierte Daten auszulesen versuchen. Datum, Aktennummern, Name des Anwalts, etc.

### Das Vorgehen
1. Wir werden dafür zuerst eine Liste erstellen aller Dokumente.
2. Alle Dokumente ein- und auslesen
3. Dann werden wir von jedem einzelnen Dokument das Datum und die Aktennummer ziehen. Und allenfalls noch den Namen des Anwalts
4. Dieses Resultat werden wir als Dictionary-Liste abspeichern
5. Dann speichern wir das als .csv ab

Um ganz einfach eine Liste von Elementen in einem Folder zu erstellen, können wir das Module ```os```verwenden. Damit wir das Module verwenden können, müssen wir das Module zuerst importieren.

#### 1. Liste aller Dokumente

````python
import os
````

````python
die_liste = os.listdir("gerichts-urteile/")
````

#### 2. Einlesen von Dokumenten

Nehmen wir das erste Dokument in der Liste:

````python
die_liste[0]
````

````python
f = open("gerichts-urteile/2C_641_2007.txt", "r")
f
````

Jetzt bauen wir eine Liste aller Text

````python
text_liste = []

for text in die_liste:
    f = open("gerichts-urteile/"+text, "r")
    text_liste.append(f.read())
````

````python
len(text_liste)
````

````python
text_liste[12]
````

#### 3.  Jetzt ziehen wir aus jedem Text das Datum und die Aktennummer

````python
import re
````

````python
datum = re.compile(r'\d{2}\.\d{2}\.\d{4}') 
d = re.search(datum, text_liste[11])
d.group()
````

````python
text_liste[9]
````

````python
aktenzeichen = re.compile(r'\d\D[_|\.]\d*/\d{4}')  #2C_641/2007 #9F_6/2015 #2A.168/2006 
a = re.search(aktenzeichen, text_liste[11])
a.group()
````

````python
advokat = re.compile(r'vertreten durch\s\w*\s\w*\s\w*')  #2C_641/2007 #9F_6/2015 #2A.168/2006 
ak = re.search(advokat, text_liste[0])
ak.group()
````

Und nun machen wir das über alle Texte.

````python
datum = re.compile(r'\d{2}\.\d{2}\.\d{4}') 
aktenzeichen = re.compile(r'\d\D[_|\.]\d*/\d{4}')
advokat = re.compile(r'vertreten durch\s\w*\s\w*\s\w*')

for elem in text_liste[:10]:
    d = re.search(datum, elem)
    a = re.search(aktenzeichen, elem)
    try:
        ak = re.search(advokat, elem)
        ak = ak.group()
    except:
        ak = "N/A"

    print(d.group(), a.group(), ak)
````

#### 4. Und nun machen wir daraus eine Dictionary-Liste

````python
datum = re.compile(r'\d{2}\.\d{2}\.\d{4}') 
aktenzeichen = re.compile(r'\d\D[_|\.]\d*/\d{4}')

dict_data = []

for elem in text_liste:
    d = re.search(datum, elem)
    a = re.search(aktenzeichen, elem)
    try:
        ak = re.search(advokat, elem)
        ak = ak.group()
    except:
        ak = "N/A"
    
    mini_dict = {'Datum': d.group(),
                 'Aktenzeichen': a.group(),
                 'Vertreter': ak}
    
    dict_data.append(mini_dict)
    
````

````python
#Fall Dauer
#Häufigkeit eines bestimmten Datums
dict_data[:10]
````

````python
len(dict_data)
````

#### 5. Und nun speichern wir das als .csv ab

````python
import csv
````

````python
csv_file = "gerichtsdokumente.csv"
csv_columns = ['Datum', 'Aktenzeichen']

try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict_data:
            writer.writerow(data)
except IOError:
    print("I/O error")
````
