```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
r = requests.get('https://www.phishtank.com/user_submissions.php?page=1&username=SWITCHCERT')
html = BeautifulSoup(r.text, 'html.parser')
total = int(html.find('h2').text.split("(")[-1].split(" ")[0].replace(",",""))
total = round(total / 20)#Finding total number of messages from Cert
total
```

```python
type(total)
```

```python
#Visiting all the SWITCHCERT mentions and saving them in to html
urlstart = 'https://www.phishtank.com/user_submissions.php?page='
urlend = '&username=SWITCHCERT'
for elem in range(1,total):
    #time.sleep(.2)
    url = urlstart + str(elem) + urlend
    print(url)
    r = requests.get(url)
    html = BeautifulSoup(r.text, 'html.parser')
    
    with open("html_switch/"+str(elem)+".htm", "w") as file:
        file.write(str(html))
        file.close()
```
