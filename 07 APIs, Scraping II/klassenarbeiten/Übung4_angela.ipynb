```python
def getcomments(url):
    
    from bs4 import BeautifulSoup
    import pandas as pd
    import requests
    
    r = requests.get(url)
    contents = r.text
    soup = BeautifulSoup(contents,"html.parser") 
    articles = soup.find_all('div', {'class':'node__content'})

    lst = []
    for article in articles:
    
        try:
            cat = article.find('a', {'class':"node__link"})['href'].split('/')[2]
            txt = article.find('a', {'class':"node__link"}).text
            lik = article.find('span', {'class':"count"}).text
    
        except:
            continue
        
    
        dct = {'Kategorie':cat, 
           'Text': txt,
           'Likes':lik}
    
        lst.append(dct)
        df = pd.DataFrame(lst)
    return df
```

```python
getcomments("https://www.tierwelt.ch/news/reportagen")
```
