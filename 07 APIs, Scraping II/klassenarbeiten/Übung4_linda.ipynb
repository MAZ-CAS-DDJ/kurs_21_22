```python
def getautoren(url):
    
    r = requests.get(url)
    contents = r.text 
    soup = BeautifulSoup(contents,"html.parser")

    titelliste = soup.find_all('div', {'class':'teaser__content'})

    nzzfrontartikel = []
    for element in titelliste:
    
        titel = element.find('span',{'class':'teaser__title-name'}).text.replace('\n', '').strip()
    
        try:
            autor = element.find('span', {'class':"metainfo__item"}).text
        except:
            autor = "N/A"
    
        minidict = {'Titel': titel,
                'Autor': autor}
    
        nzzfrontartikel.append(minidict)
    
    df = pd.DataFrame(nzzfrontartikel)
    
    return df
```

```python
getautoren("https://www.nzz.ch/international")
```
