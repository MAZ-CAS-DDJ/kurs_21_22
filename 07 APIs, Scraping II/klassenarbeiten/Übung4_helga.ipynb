```python
def TrumpQuotient(url):
    # Wichtige Packages importieren
    import requests
    from bs4 import BeautifulSoup
    import pandas as pd
    import time
    
    # Informationen von der Website holen
    wp = requests.get(url, time.sleep(10))
    wp_contents = wp.text # Wir lesen den Inhalt aus...
    soup = BeautifulSoup(wp_contents,"html.parser")
    
    titelliste = soup.find_all('h2') #...und suchen nach Überschriften auf Artikelebene
    # Jetzt erstellen wir eine Liste der Trump-Treffer
    
    hitlist = [] 
    for t in titelliste:
        t = t.text # Erst in text umwandeln
        treffer = t.find("Trump") 
        if treffer >= 0: 
            treffer = "Yes"
        else:
            treffer = "No"
        hitlist.append(treffer) # dies wg der Ausgabe von find() in integers
    # Die Liste mit den Titeln und die mit den Treffern werden zu einem Dataframe zusammengefügt
    
    tq_database = []
    for t,h in zip(titelliste,hitlist):
        eintrag= {'Titel:':t.text,
                  'Trump-Treffer:':h} 
        tq_database.append(eintrag)
    
    # Ausgabe der Ergebnisse in drei Textzeilen     
    trumpquotient = hitlist.count("Yes")/len(hitlist) * 100
    print("Auf " + url + " gibt es aktuell " + str(len(hitlist)) + " Artikel.")
    print(str(hitlist.count("Yes")) + " davon enthalten den Namen Trump.")
    print("Der Trump-Quotient liegt damit bei " + str(round(trumpquotient,1)) + " Prozent.")
    
    # Optional: Artikelliste anzeigen
    return pd.DataFrame(tq_database)[1:]
```

```python
TrumpQuotient("https://www.nytimes.com")
```
