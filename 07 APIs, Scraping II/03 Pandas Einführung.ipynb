## Daten direkt importieren und visualisieren

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
country = input("Enter the country you want to look up?")
```

```python
ff = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
ff = ff[ff['Country/Region']==country] #Land ändern
ff = ff.transpose()
ff = ff[12:]
ff['sum'] = ff.sum(axis=1)
ff['sum diff'] = ff['sum'].diff()
ff['rolling'] = ff['sum diff'].rolling(13).mean()

ef = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
ef = ef[ef['Country/Region']==country ] #Land ändern
ef = ef.transpose()
ef = ef[12:]
ef['sum'] = ef.sum(axis=1)
ef['sum diff'] = ef['sum'].diff()
ef['rolling'] = ef['sum diff'].rolling(13).mean()

ff_ = ff.merge(ef, left_index=True, right_index=True)

ff_['normalized cases']=ff_['rolling_x'] / ff_['rolling_x'].max()
ff_['normalized dead']=ff_['rolling_y'] / ff_['rolling_y'].max()

ff_[["normalized cases", "normalized dead"]].copy().plot()
```
