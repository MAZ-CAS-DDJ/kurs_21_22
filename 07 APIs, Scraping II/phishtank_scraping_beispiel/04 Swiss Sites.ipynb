## Finding Swiss sites

```python
def ch(elem):
    if ".ch" in elem:
        return True
    else:
        return False
    
ph['ch check'] = ph['URL'].apply(ch)
ch = ph[ph['ch check']==True]
```

```python
ch.info()
```

Finding Top Level Domain

```python
def toplevel(elem):
    elem = elem.replace("https://","").replace("http://","").split("/")[0]
    return elem
```

```python
ch['toplevel'] = ch['URL'].apply(toplevel)
```

```python
ch
```

```python
test = "https://secureclients.ubs.ch"
```

```python
def endinginch(elem):
    if elem[-3:] == '.ch':
        return True
    else: 
        return False
```

```python
ch['ch toplevel'] = ch['toplevel'].apply(endinginch)
```

# Examining Swiss toplevel domains

```python
chtl = ch[ch['ch toplevel']==True]
```
